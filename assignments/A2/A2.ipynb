# Exercise 1

```python
s = symbols('s', real=True)
```

## a

```python
gamma = Matrix([s, s*s, s*s*s])
T = diff(gamma, s)
T = T / T.norm()
T
```

## b

```python
dT= diff(T, s)
N = dT / dT.norm()

N
```

```python
kappa =-N.dot(dT)
kappa
```

## c

```python
B = T.cross(N)
B
```

```python
dB = diff(B, s)
tau = N.dot(dB)
tau
```
