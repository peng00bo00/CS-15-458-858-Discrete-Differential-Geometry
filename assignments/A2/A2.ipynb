# Exercise 1

```python
s = symbols('s', real=True)
```

## a

```python
gamma = Matrix([s, s*s, s*s*s])
T = diff(gamma, s)
T = T / T.norm()
T
```

## b

```python
dT= diff(T, s)
N = dT / dT.norm()

N
```

```python
kappa =-N.dot(dT)
kappa
```

## c

```python
B = T.cross(N)
B
```

```python
dB = diff(B, s)
tau = N.dot(dB)
tau
```

# Exercise 2

```python
u, v = symbols('u v', real=True)
```

```python
x = (2 * u) / (u*u + v*v + 1)
y = (2 * v) / (u*u + v*v + 1)
z = (u*u + v*v - 1) / (u*u + v*v + 1)

f = Matrix([x, y, z])
f
```

## b

```python
df = f.jacobian(Matrix([u,v]))
df
```

## c

```python
g = df.T * df
g
```

## d

```python
N = df[:, 0].cross(df[:, 1])
N = N / N.norm()
N
```

## e

```python
dN = N.jacobian(Matrix([u,v]))
dN
```
