```python
import numpy as np
from scipy import linalg
```

```python
V = 5

## cotan
cot = 1.0 / np.tan(np.pi/3)
## edge length
l = 1.0
## triangle area
area = 0.5 * l * l * np.sin(np.pi/3)

## adjacent matrix
Adj = np.zeros((V, V))

Adj[0, 1] = 1
Adj[0, 2] = 1
Adj[0, 3] = 1
Adj[0, 4] = 1

Adj[1, 2] = 1
Adj[1, 4] = 1

Adj[2, 3] = 1
Adj[2, 4] = 1

Adj[3, 4] = 1

Adj = Adj + Adj.T

## cotan Laplacian
L = np.zeros((V, V))

for i in range(V):
  s = 1e-8
  for j in range(V):
    if Adj[i, j] > 0:
      L[i, j] = -cot
      s += cot
  L[i, i] = s

L = -L

## M
M = np.zeros((V, V))

for i in range(V):
  M[i, i] = Adj[i].sum() * area / 3

## f
f = np.zeros(V)

f[0] = 1.0
f[1] = 0.0
f[2] = -130/33
f[3] = 108/11
f[4] = -130/33
```

```python
Lii = L[:2, :2]
Lib = L[:2, 2:]
Lbi = L[2:, :2]
Lbb = L[2:, 2:]

Mii = M[:2, :2]
Mbb = M[2:, 2:]

fii = f[:2]
fbb = f[:2]
```

```python
phib = np.array([1, -2, 1])
phii = linalg.solve(Lii, Mii @ fii - Lib @ phib)

phi = np.hstack((phii, phib))
phi
```
